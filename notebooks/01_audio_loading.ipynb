# 01 — Audio Loading

Load audio files, inspect their properties, and play them back.

```python
from pathlib import Path
from IPython.display import Audio, display

from rytmi import load_audio, trim_silence, normalize
```

```python
# Load the synthetic click track
sample_path = Path("../data/samples/click_120bpm.wav")
audio = load_audio(sample_path)

print(f"Sample rate: {audio.sr} Hz")
print(f"Duration:    {audio.duration:.2f} s")
print(f"Samples:     {len(audio.samples)}")
print(f"Dtype:       {audio.samples.dtype}")
print(f"Range:       [{audio.samples.min():.3f}, {audio.samples.max():.3f}]")
```

```python
# Play the audio in the notebook
display(Audio(audio.samples, rate=audio.sr))
```

```python
# Trim silence and normalize
trimmed = trim_silence(audio)
normed = normalize(trimmed)

print(f"Original duration: {audio.duration:.2f} s")
print(f"Trimmed duration:  {trimmed.duration:.2f} s")
print(f"Normalized range:  [{normed.samples.min():.3f}, {normed.samples.max():.3f}]")
```

```python
# Load with different sample rate
audio_44k = load_audio(sample_path, sr=44100)
print(f"44.1kHz: {len(audio_44k.samples)} samples, {audio_44k.duration:.2f}s")
```
