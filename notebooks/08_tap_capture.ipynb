# Phase 28 P2 — Tap capture

Tap along to each track. Aim for **at least three takes per track** so the consensus utility (Phase 28 P3) can drop outliers.

Per take:

1. Run the cell — audio player + TAP / Save / Reset buttons appear.
2. Press Play in the audio widget.
3. Click the TAP button on every perceived beat. (Tip: focus it once with Tab; then spacebar acts as click — the browser binds Space to the focused button.)
4. Click **Save take**. The recorder picks the next free slot automatically (`take_1.json`, then `take_2.json`, …) and clears the buffer so you can immediately tap a new take in the same cell.
5. Click **Reset** instead of Save to discard the current buffer without writing.

Track suggestions: stick to ones where you can clearly hear the batida. If a track feels ambiguous (Bonga / Mona Ki Ngi Xica was that case), skip it — guessed taps are worse than missing taps.
