# 02 — Onset & Beat Detection

Run the DSP pipeline: detect onsets, track beats, estimate tempo.

```python
from pathlib import Path
import numpy as np
from IPython.display import Audio, display

from rytmi import load_audio, analyze
from rytmi.dsp import detect_onsets, track_beats, estimate_tempo
```

```python
audio = load_audio(Path("../data/samples/click_120bpm.wav"))
display(Audio(audio.samples, rate=audio.sr))
```

```python
# Detect onsets
onsets = detect_onsets(audio)
print(f"Detected {len(onsets.times)} onsets")
print(f"Onset times (s): {np.round(onsets.times, 3)}")
```

```python
# Track beats
beats = track_beats(audio)
print(f"Tempo: {beats.tempo:.1f} BPM")
print(f"Detected {len(beats.times)} beats")
print(f"Beat times (s): {np.round(beats.times, 3)}")
```

```python
# Standalone tempo estimation
tempo = estimate_tempo(audio)
print(f"Estimated tempo: {tempo:.1f} BPM")
```

```python
# Full analysis pipeline
result = analyze(audio)

print(f"Tempo:    {result.tempo:.1f} BPM")
print(f"Onsets:   {len(result.onsets.times)}")
print(f"Beats:    {len(result.beats.times)}")
if result.inter_onset_intervals is not None:
    print(f"Mean IOI: {np.mean(result.inter_onset_intervals):.1f} ms")
    print(f"Std IOI:  {np.std(result.inter_onset_intervals):.1f} ms")
```
