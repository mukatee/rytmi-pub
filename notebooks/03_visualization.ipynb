# 03 — Visualization

Visualize audio analysis: waveform, spectrogram, onsets, beats.

```python
%matplotlib inline
from pathlib import Path
import matplotlib.pyplot as plt

from rytmi import load_audio, analyze, plot_analysis
from rytmi.viz import plot_waveform, plot_spectrogram, plot_onsets, plot_beats, plot_onset_strength
```

```python
audio = load_audio(Path("../data/samples/click_120bpm.wav"))
result = analyze(audio)
print(f"Tempo: {result.tempo:.1f} BPM | Onsets: {len(result.onsets.times)} | Beats: {len(result.beats.times)}")
```

## Individual plots

```python
plot_waveform(audio)
plt.show()
```

```python
plot_spectrogram(audio)
plt.show()
```

```python
plot_onsets(audio, result.onsets)
plt.show()
```

```python
plot_beats(audio, result.beats)
plt.show()
```

```python
plot_onset_strength(result.onsets)
plt.show()
```

## Full analysis overview

```python
fig = plot_analysis(result)
plt.show()
```
